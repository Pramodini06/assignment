```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('/content/drive/MyDrive/new/Default_Fin.csv')
```

```python
data.head(5)
```

```python
sns.set_style('whitegrid')
data['Bank Balance'].hist(bins=30)
plt.xlabel('Bank Balance')
```

```python
X = data[['Index', 'Employed', 'Bank Balance','Annual Salary']]
y = data['Defaulted?']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
data.shape
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
logistic_model = LogisticRegression(max_iter=500)
logistic_model.fit(X_train,y_train)
```

```python
predictions = logistic_model.predict(X_test)
```

```python
predictions
```

```python
logistic_model.score(X_test,y_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
report = classification_report(y_test, predictions, output_dict=True)
df = pd.DataFrame(report).transpose()

plt.figure(figsize=(10, 6))
sns.heatmap(df.iloc[:-1, :-1], annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Classification Metrics Heatmap')
plt.tight_layout()
plt.show()
```

Now, let's train an SVM model and make predictions.

```python
from sklearn.svm import SVC

# Initialize and train the SVM model
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Make predictions
svm_predictions = svm_model.predict(X_test)
```

Let's evaluate the performance of the SVM model using a classification report and visualize it.

```python
print(classification_report(y_test, svm_predictions))

report_svm = classification_report(y_test, svm_predictions, output_dict=True)
df_svm = pd.DataFrame(report_svm).transpose()

plt.figure(figsize=(10, 6))
sns.heatmap(df_svm.iloc[:-1, :-1], annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('SVM Classification Metrics Heatmap')
plt.tight_layout()
plt.show()
```
